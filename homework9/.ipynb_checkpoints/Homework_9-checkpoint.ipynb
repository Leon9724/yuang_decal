```python
# My Jupiter notebook keeps bouncing back n force when I reach the bottom of the page, so I have to make multiple notebooks.
# 1.1
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('GlobalLandTemperaturesByState.csv')
print(df)
```

```python
df = df[['dt', 'AverageTemperature', 'State']]
df['dt'] = pd.to_datetime(df['dt'])
df = df[df['dt'].dt.year > 2000]

states_filtered = ['Wyoming', 'Nebraska', 'South Dakota']
df_filtered = df[df['State'].isin(states_filtered)]

print(df_filtered.shape)  
print(df_filtered.head())
```

```python
# 1.2

df_avg_temp_by_date = df_filtered.groupby('dt')['AverageTemperature'].mean().reset_index()
df_avg_temp_by_date.columns = ['Date', 'AverageTemperature']
print(df_avg_temp_by_date.head())


```

```python
# 1.3
plt.figure(figsize=(12, 6))
plt.plot(df_avg_temp_by_date['Date'], df_avg_temp_by_date['AverageTemperature'],
         marker='o', linestyle='-', markersize=1)

plt.xlabel('Date')
plt.ylabel('Average Temperature (°C)')
plt.title('Average Temperature Over Time (WY, NE, SD)')
plt.xticks(rotation=45)
plt.show()
```
